```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('data_end.npz')
high = data['arr_0'] + data['arr_1']
low = data['arr_0'] - data['arr_1']
```

```python

ind = 2

data2 = np.load('../../compensator_plots/InvertedPendulumModified-multi-v102018-06-11T15:45:03/data_4.0.npz')
high2 = (data2['arr_0'] + data2['arr_1'])[:, ind]
low2 = (data2['arr_0'] - data2['arr_1'])[:, ind]
rwd2 = data2['arr_0'][:, ind]
```

```python
time_step_inc = 2048

fig = plt.figure(figsize=(20, 8))

subplot_ax = fig.add_subplot(1, 1, 1)

subplot_ax.set_title('InvertedPendulumMulti Trained on Weighted Ensemble of Compensator Policies', fontsize = 24)
subplot_ax.plot(data['arr_0'], color='r',linewidth=4)
subplot_ax.set_xlabel('{} Time Steps'.format(time_step_inc), fontsize=24)
subplot_ax.set_ylabel('Average Reward', fontsize=24)
```

```python
time_step_inc = 2048

fig = plt.figure(figsize=(20, 8))

subplot_ax = fig.add_subplot(1, 1, 1)

subplot_ax.set_title('InvertedPendulumMulti Trained on Weighted Ensemble of Compensator Policies', fontsize = 24)
subplot_ax.plot(np.arange(len(data['arr_0'])), data['arr_0'],
                np.arange(len(data['arr_0'])), high,
                np.arange(len(data['arr_0'])), low,
                color='r',linewidth=4)
subplot_ax.fill_between(np.arange(len(data['arr_0'])), high.squeeze(), low.squeeze(), color='r',linewidth=4, alpha=0.5)
subplot_ax.set_xlabel('{} Time Steps'.format(time_step_inc), fontsize=24)
subplot_ax.set_ylabel('Average Reward', fontsize=24)
```
